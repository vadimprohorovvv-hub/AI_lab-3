```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)
y1 = x
y2 = x ** 2
y3 = x ** 3

plt.figure(figsize=(8, 5))

plt.plot(x, y1, color='blue', linestyle='--', linewidth=2, label='$y = x$')
plt.plot(x, y2, color='green', linestyle='-.', linewidth=2, label='$y = x^2$')
plt.plot(x, y3, color='red', linestyle='-', linewidth=2, label='$y = x^3$')

plt.title('Три функции на одном графике', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)
plt.legend(loc='upper left', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

__Задание 5__ Столбчатая диаграмма
<p>Создайте столбчатую диаграмму, которая показывает количество студентов, получивших оценки</p>

```python

categories = ['Отлично', 'Хорошо', 'Удовлетворительно', 'Неудовлетворительно']
counts = [20, 35, 30, 15]
colors = ['gold', 'lightgreen', 'lightskyblue', 'lightcoral']

# Построение столбчатой диаграммы
plt.figure(figsize=(8, 6))

bars = plt.bar(categories, counts, color=colors, edgecolor='black', linewidth=1.5)

# Добавление значений на столбцах
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{int(height)}', ha='center', va='bottom', fontsize=11)

# Оформление
plt.title('Распределение оценок студентов', fontsize=16, fontweight='bold')
plt.xlabel('Оценка', fontsize=12)
plt.ylabel('Количество студентов', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3, axis='y')  # сетка только по Y

# Настройка пределов оси Y для отображения подписей
plt.ylim(0, max(counts) + 5)

plt.show()

# Альтернативный вариант: горизонтальная столбчатая диаграмма
plt.figure(figsize=(8, 5))
plt.barh(categories, counts, color=colors, edgecolor='black', linewidth=1.5)
plt.title('Распределение оценок студентов (горизонтальная)', fontsize=14)
plt.xlabel('Количество студентов', fontsize=12)
plt.ylabel('Оценка', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3, axis='x')
plt.show()
```

__Задание 6__ Круговая диаграмма
<p>Используя данные предыдущей задачи, постройте круговую диаграмму с процентными подписями секторов.</p>

```python

categories = ['Отлично', 'Хорошо', 'Удовлетворительно', 'Неудовлетворительно']
counts = [20, 35, 30, 15]
colors = ['gold', 'lightgreen', 'lightskyblue', 'lightcoral']

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))

# explode - выделение секторов (чуть отодвигаем "Отлично" и "Неудовлетворительно")
explode = (0.05, 0, 0, 0.05)

wedges, texts, autotexts = plt.pie(counts, 
                                     labels=categories,
                                     colors=colors,
                                     autopct='%1.1f%%',  # процентные подписи с одним знаком после запятой
                                     explode=explode,
                                     startangle=90,       # начальный угол (90 градусов - от верха)
                                     shadow=True,         # тень
                                     textprops={'fontsize': 12})

# Настройка подписей
for text in texts:
    text.set_fontweight('bold')
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)
plt.axis('equal')  # обеспечение круглой формы
plt.tight_layout()
plt.show()
```

__Задание 7__ Трехмерный график поверхности
<p>Используя mpl_toolkits.mplot3d, постройте 3D-график функции</p>

```python

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, 
                       cmap='viridis',      
                       linewidth=0,         
                       antialiased=True,    
                       alpha=0.9)
ax.set_xlabel('Ось X', fontsize=12, labelpad=10)
ax.set_ylabel('Ось Y', fontsize=12, labelpad=10)
ax.set_zlabel('Ось Z', fontsize=12, labelpad=10)

ax.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()
```

__Задание 8__ Множественные подграфики
<p>Постройте четыре графика в одной фигуре</p>

```python

x = np.linspace(-10, 10, 200)
y1 = x           # y = x
y2 = x ** 2      # y = x^2

# Создание фигуры с 4 подграфиками (2×2)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Линейный график y = x
axes[0, 0].plot(x, y1, color='blue', linewidth=2)
axes[0, 0].set_title('Линейный график: $y = x$', fontsize=12)
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')
axes[0, 0].grid(True, linestyle='--', alpha=0.7)
axes[0, 0].axhline(0, color='black', linewidth=0.5)
axes[0, 0].axvline(0, color='black', linewidth=0.5)

# 2. Парабола y = x^2
axes[0, 1].plot(x, y2, color='green', linewidth=2)
axes[0, 1].set_title('Парабола: $y = x^2$', fontsize=12)
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')
axes[0, 1].grid(True, linestyle='--', alpha=0.7)
axes[0, 1].axhline(0, color='black', linewidth=0.5)
axes[0, 1].axvline(0, color='black', linewidth=0.5)

# 3. Кубическая функция y = x^3 (дополнительно)
x_cube = np.linspace(-3, 3, 200)  # ограничим диапазон для лучшей визуализации
y3 = x_cube ** 3
axes[1, 0].plot(x_cube, y3, color='red', linewidth=2)
axes[1, 0].set_title('Кубическая функция: $y = x^3$', fontsize=12)
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].grid(True, linestyle='--', alpha=0.7)
axes[1, 0].axhline(0, color='black', linewidth=0.5)
axes[1, 0].axvline(0, color='black', linewidth=0.5)

# 4. Синусоида y = sin(x) (дополнительно)
x_sin = np.linspace(-2*np.pi, 2*np.pi, 200)
y4 = np.sin(x_sin)
axes[1, 1].plot(x_sin, y4, color='purple', linewidth=2)
axes[1, 1].set_title('Синусоида: $y = \sin(x)$', fontsize=12)
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y')
axes[1, 1].grid(True, linestyle='--', alpha=0.7)
axes[1, 1].axhline(0, color='black', linewidth=0.5)
axes[1, 1].axvline(0, color='black', linewidth=0.5)

# Общий заголовок для всей фигуры
fig.suptitle('Множественные подграфики (2×2 сетка)', fontsize=16, fontweight='bold', y=0.98)

# Автоматическая настройка отступов
plt.tight_layout()
plt.show()
```

__Задание 3__ Площадь под логарифмической кривой
<p>Вычислите площадь под графиком функции</p>

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.log(x)

a, b = 1, 4

n = 1000 
x_trap = np.linspace(a, b, n)
y_trap = f(x_trap)
integral_trap = np.trapz(y_trap, x_trap) 

exact = (b * np.log(b) - b) - (a * np.log(a) - a)

x = np.linspace(0.5, 4.5, 1000)
y = f(x)

plt.figure(figsize=(12, 7))

plt.plot(x, y, 'b-', linewidth=3, label='$f(x) = \ln(x)$')

x_fill = np.linspace(a, b, 500)
y_fill = f(x_fill)
plt.fill_between(x_fill, y_fill, 0, alpha=0.4, color='lightgreen',
                  label=f'Площадь ≈ {integral_trap:.4f}')

plt.axvline(x=a, color='red', linestyle='--', alpha=0.7)
plt.axvline(x=b, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=0, color='black', linewidth=1)

plt.title('Площадь под графиком $f(x) = \ln(x)$', fontsize=16, fontweight='bold')
plt.xlabel('x', fontsize=13)
plt.ylabel('f(x)', fontsize=13)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend(loc='upper left', fontsize=12)

plt.text(2.2, 1.2, f'$\\int_{1}^{4} \\ln(x) dx = {integral_trap:.6f}$\n(метод трапеций)\nТочное значение: {exact:.6f}',
         fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.xlim(0.5, 4.5)
plt.ylim(-0.5, 1.6)
plt.tight_layout()
plt.show()

print(f"Интеграл (метод трапеций): {integral_trap:.6f}")
print(f"Интеграл (точное значение): {exact:.6f}")
```

__Задание 4__ Волны на воде
<p>Постройте график функции: f(x, y) = sin(x) * sin(y)</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm', linewidth=0, antialiased=True)

ax.set_title('Волны на воде: $f(x,y) = \sin(x) \cos(y)$', fontsize=14)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.tight_layout()
plt.show()
```
